```python
import pandas as pd
```

```python
df = pd.read_excel("nfl.xlsx")
```

```python
df.set_index("Date", inplace=True)
```

```python
years = df.loc["2017-09-10":"2020"]
```

```python
san_fran = years.loc[(years["Home Team"] == "San Francisco 49ers") | (years["Away Team"] == "San Francisco 49ers")]
san_fran = san_fran[["Home Team", "Away Team", "Home Score", "Away Score", "Total Score Open"]]
```

```python
san_fran["Home Team"].loc["2019-10-20"] = "Washington Football Team"
```

```python
san_fran["Home Team"].loc["2017-10-15"] = "Washington Football Team"
```

```python
san_fran["Home Team"]
```

```python
stadium = pd.read_html("https://en.wikipedia.org/wiki/List_of_current_National_Football_League_stadiums")
```

```python
stadium = stadium[1]
```

```python
stadium = stadium.drop(columns=(["Image", "Capacity", "Opened", "Ref(s)"]))
```

```python
stadium = stadium.append(stadium.iloc[[20,25]])
```

```python
stadium.reset_index(inplace=True,drop=True)
```

```python
stadium["Team(s)"].loc[20] = "New York Giants"
stadium["Team(s)"].loc[30] = "New York Jets"
stadium["Team(s)"].loc[25] = "Los Angeles Rams"
stadium["Team(s)"].loc[31] = "Los Angeles Chargers"
stadium 
```

```python
SF_joined = san_fran.join(stadium.set_index("Team(s)"), on = "Home Team")
```
